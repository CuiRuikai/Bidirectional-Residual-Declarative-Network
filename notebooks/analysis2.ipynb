# Ablation Study 1 -- Bidirectional FC Layer Performance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_df={}
proj_type=['L1S','L1B','L2S','L2B','LInfS','LInfB']

# data_df['None']=pd.read_csv('csv_7class/no.csv',index_col=0)
data_df['L1S']=pd.read_csv('csv_bdnn/l1s.csv',index_col=0)
data_df['L1B']=pd.read_csv('csv_bdnn/l1b.csv',index_col=0)
data_df['L2S']=pd.read_csv('csv_bdnn/l2s.csv',index_col=0)
data_df['L2B']=pd.read_csv('csv_bdnn/l2b.csv',index_col=0)
data_df['LInfS']=pd.read_csv('csv_bdnn/linfs.csv',index_col=0)
data_df['LInfB']=pd.read_csv('csv_bdnn/linfb.csv',index_col=0)
data_df['L1S'].head()
```

```python
for i in proj_type:
    plt.plot(data_df[i]['acc@1'], label=i)

plt.ylabel('top1 acc (%)')
plt.xlabel('epoch')
plt.legend()
plt.show()
```

```python
for i in proj_type:
    plt.plot(data_df[i]['acc@3'], label=i)

plt.ylabel('top1 acc (%)')
plt.xlabel('epoch')
plt.legend()
plt.show()
```
