# Ablation Study 1 -- Projection Layer Performance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_df={}
proj_type=['None','L1S','L1B','L2S','L2B','LInfS','LInfB']

data_df['None']=pd.read_csv('csv_7class/no.csv',index_col=0)
data_df['L1S']=pd.read_csv('csv_7class/l1s.csv',index_col=0)
data_df['L1B']=pd.read_csv('csv_7class/l1b.csv',index_col=0)
data_df['L2S']=pd.read_csv('csv_7class/l2s.csv',index_col=0)
data_df['L2B']=pd.read_csv('csv_7class/l2b.csv',index_col=0)
data_df['LInfS']=pd.read_csv('csv_7class/linfs.csv',index_col=0)
data_df['LInfB']=pd.read_csv('csv_7class/linfb.csv',index_col=0)
data_df['None'].head()
```

```python
fig=plt.figure()
```

## acc@1

```python
for i in proj_type:
    plt.plot(data_df[i]['acc@1'], label=i)

plt.ylabel('top1 acc (%)')
plt.xlabel('epoch')
plt.legend()
plt.savefig('figures/acc1.eps', format='eps')
plt.show()
```

```python
for i in proj_type:
    plt.plot(data_df[i]['acc@3'], label=i)

plt.ylabel('top3 acc (%)')
plt.xlabel('epoch')
plt.legend()
plt.savefig('figures/acc3.eps', format='eps')
plt.show()
```

```python
for i in proj_type:
    plt.plot(data_df[i]['mAP'], label=i)

plt.ylabel('mAP')
plt.xlabel('epoch')
plt.legend()
plt.savefig('figures/map.eps', format='eps')
plt.show()
```

```python
for i in proj_type:
    plt.plot(data_df[i]['loss'], label=i)

plt.ylim(1,4)
plt.ylabel('top1 acc (%)')
plt.xlabel('epoch')
plt.legend()
plt.show()
```
